```python
import numpy as np
import lal
import lalpulsar
from pycbc.detector import Detector
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

def phase_evolution(t, f0=100.0, f1dot=1e-9, f2dot=0.0, f3dot=0.0, f4dot=0.0, phi0=0.0):
    """
    Compute the phase evolution of a continuous GW signal including up to 4th order derivatives.

    Parameters:
        t (float or array): Time array in seconds
        f0 (float): Initial frequency at reference time
        f1dot (float): First frequency derivative (Hz/s)
        f2dot (float): Second frequency derivative (Hz/s²)
        f3dot (float): Third frequency derivative (Hz/s³)
        f4dot (float): Fourth frequency derivative (Hz/s⁴)
        phi0 (float): Initial phase (radians)

    Returns:
        phase (array): Computed phase at each time step
    """
    return phi0 + 2 * np.pi * (
        f0 * t
        + 0.5 * f1dot * t**2
        + (1./6) * f2dot * t**3
        + (1./24) * f3dot * t**4
        + (1./120) * f4dot * t**5
    )

def compute_antenna_patterns(detector, alpha, delta, psi, t):
    """
    Compute the detector response functions (F_+, F_x) for a given sky location and polarization.

    Parameters:
        detector (str): LIGO/Virgo detector name ('H1', 'L1', 'V1', etc.)
        alpha (float): Right ascension (radians)
        delta (float): Declination (radians)
        psi (float): Polarization angle (radians)
        tref (float): Reference GPS time

    Returns:
        F_plus, F_cross (floats): Detector response factors
    """
    det = Detector(detector)
    fp, fc = det.antenna_pattern(alpha, delta, psi, t)
    
    return fp, fc

def waveform(
    h0, cosi, f0, f1dot, f2dot, f3dot, f4dot, 
    tstart, tref, duration, fs, phi0, alpha, delta, psi, detector
):
    """
    Generate a continuous gravitational wave signal with explicit phase control and antenna patterns.

    Parameters:
        h0 (float): GW amplitude
        cosi (float): Cosine of inclination angle
        f0 (float): Initial frequency (Hz)
        f1dot, f2dot, f3dot, f4dot (float): Frequency derivatives (Hz/s, Hz/s², Hz/s³, Hz/s⁴)
        tref (float): Reference time (GPS)
        duration (float): Total duration of the signal (seconds)
        dt (float): Time step (seconds)
        phi0 (float): Initial phase (radians)
        alpha (float): Right ascension (radians)
        delta (float): Declination (radians)
        psi (float): Polarization angle (radians)
        detector (str): LIGO/Virgo detector name ('H1', 'L1', etc.)

    Returns:
        times (array): Time array
        h_plus (array): Plus-polarization waveform
        h_cross (array): Cross-polarization waveform
        h_detector (array): Projected strain on the detector
    """
    t = np.linspace(tstart, tstart+duration, int(duration*fs))
    print(t.shape)
    phase = phase_evolution(t-tref, f0, f1dot, f2dot, f3dot, f4dot, phi0)

    # Compute antenna pattern functions
    Fplus, Fcross = compute_antenna_patterns(detector, alpha, delta, psi, t)

    # Compute GW amplitudes
    ap = h0 * (1 + cosi**2) / 2
    ax = h0 * cosi

    # Compute GW polarizations
    h_plus = ap * np.cos(phase)
    h_cross = ax * np.sin(phase)

    # Project onto detector response
    h_detector = Fplus * h_plus + Fcross * h_cross

    return t, h_plus, h_cross, h_detector
```

```python

def compute_fft_segments(times, signal, f0, band, segment_length=1800, fs=1024):
    """
    Computes the FFT of the signal in segments and returns the portion around f0.

    Parameters:
    - times: Array of time values.
    - signal: GW strain signal.
    - f0: Central frequency of interest (Hz).
    - band: Frequency band around f0 to keep (Hz).
    - segment_length: Length of each FFT segment in seconds.
    - fs: Sampling frequency (Hz).

    Returns:
    - segment_times: Array of start times for each segment.
    - freqs_selected: Selected frequency bins within [f0-band, f0+band].
    - fft_data_selected: 2D array of complex Fourier coefficients within selected range.
    """
    segment_samples = segment_length * fs
    num_segments = len(times) // segment_samples
    # Store results
    segment_times = []
    fft_data_selected = []
    
    for i in range(num_segments):
        start_idx = i * segment_samples
        end_idx = start_idx + segment_samples

        segment_time = times[start_idx:end_idx]
        segment_signal = signal[start_idx:end_idx]

        # Perform FFT
        fft_values = fft(segment_signal)
        freqs = fftfreq(segment_samples, d=1./fs)
        
        print(freqs.min())
        print(freqs.max())
        
        # Select frequencies within [f0 - band, f0 + band]
        freq_mask = (freqs >= f0 - band) & (freqs <= f0 + band)
        freqs_selected = freqs[freq_mask]
        fft_values_selected = fft_values[freq_mask]

        # Store segment results
        segment_times.append(segment_time[0])  # Start time of the segment
        fft_data_selected.append(fft_values_selected)  # Store selected FFT data

    return np.array(segment_times), freqs_selected, np.array(fft_data_selected)

# Example Usage
# segment_times, freqs_selected, fft_data_selected = compute_fft_segments(times, h_detector, f0=10.0, band=0.1, segment_length=1800, fs=1)
```

```python
import os

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

%matplotlib inline

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.size"] = 20

def plot_spectrograms(timestamps, frequency, fourier_data):
    fig, axs = plt.subplots(1, 1, figsize=(16, 10))
    axs = [axs]
    for ax in axs:
        ax.set(xlabel="Days", ylabel="Frequency [Hz]")

    time_in_days = (timestamps - timestamps[0]) / 86400

    axs[0].set_title("H1")
    c = axs[0].pcolormesh(
        time_in_days,
        frequency,
        fourier_data,
        norm=colors.CenteredNorm(),
    )
    fig.colorbar(c, ax=axs[0], orientation="horizontal", label="Fourier Amplitude")
    return fig, axs


plot_spectrograms(segment_times, freqs_pos, np.abs(fft_data).T)
```
